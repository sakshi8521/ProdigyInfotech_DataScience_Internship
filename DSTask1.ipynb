```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
# Set a nice style for your charts
sns.set_theme(style="whitegrid")
```

```python
# This line ensures your plots are displayed directly in the notebook
%matplotlib inline
```

```python
df = pd.read_csv("TimelineData.csv", skiprows=4)
metadata = pd.read_csv('Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_127039.csv')
# Merge the main data with the metadata on the Country Code
merged_df = pd.merge(df, metadata, on='Country Code')
df.head()
```

#### Categorical Visualization: Bar Chart

```python
# Plot the distribution of regions
plt.figure(figsize=(12, 6))
sns.countplot(data=merged_df, y='Region', palette='viridis', 
              order=merged_df['Region'].value_counts().index)
plt.title('Distribution of Countries by Region')
plt.xlabel('Number of Countries')
plt.ylabel('Region')
plt.show()
```

##### Insights : The Pareto distribution of the metadata reveals that approximately 25% of the world's reporting entities are located in Europe & Central Asia, making it the most granular region in this dataset. Conversely, North America represents the smallest category, indicating that while it is a major economic driver, it provides the least amount of individual data points for categorical comparison.

#### Continuous Visualization: Histogram

```python
# Select the year 2023 and drop any empty rows
pop_data = df['2023'].dropna()

plt.figure(figsize=(10, 6))
sns.histplot(pop_data, bins=30, kde=True, color='Pink')

# Using a log scale makes the data easier to read
plt.xscale('log') 
plt.title('Distribution of World Population (2023)')
plt.xlabel('Population Size (Log Scale)')
plt.ylabel('Frequency')
plt.show()
```

##### The chart shows that world population is extremely lopsided. While a few giant countries like India and China have huge populations, most countries in the world actually have fewer than 20 million people. This tells us that the 'average' population size is misleading because it's being pushed up by just a few massive nations, while the majority of the world's countries are actually quite small in comparison
